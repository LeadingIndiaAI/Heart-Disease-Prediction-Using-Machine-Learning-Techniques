```python
df.drop('patient_id',axis=1,inplace=True)
```

```python
df = pd.get_dummies(df, columns = ['slope','thal','cp','ves','fbs','rese','sex','ang'])
```

```python
target = pd.read_csv(r'C:\Users\Shardul\Desktop\Machine\new\train_labels.csv',sep=',')
```

```python
target.head()
```

```python
df.info()
```

```python
y = target['heart_disease_present']
X= df
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
import warnings
```

```python
classifier = Sequential()
classifier.add(Dense(128, activation = 'relu', input_dim = 28))
classifier.add(Dropout(0.5))
classifier.add(Dense(128, activation = 'relu'))
classifier.add(Dropout(0.6))
classifier.add(Dense(128, activation = 'relu'))
classifier.add(Dropout(0.5))
classifier.add(Dense(1,  activation = 'sigmoid'))
classifier.compile(optimizer = 'RMSprop', loss = 'binary_crossentropy', metrics = ['accuracy',keras_metrics.recall(),keras_metrics.precision()])
```

```python
classifier.fit(X_train, y_train, validation_data=[X_test,y_test], epochs = 20)
```

```python
from sklearn.metrics import log_loss
clf_probs=classifier.predict_proba(X_test)
score=log_loss(y_test,clf_probs)
print(score)
```
